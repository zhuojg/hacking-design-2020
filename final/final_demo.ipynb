```python
import pyimgsaliency as psal
from PIL import Image
import matplotlib.pyplot as plt

img_path = '../data/image_4_3.jpg'

plt.figure()
plt.imshow(Image.open(img_path))
plt.show()
```

```python
mbd = psal.get_saliency_mbd(img_path).astype('uint8')
binary_sal = psal.binarise_saliency_map(mbd, method='adaptive')

plt.rcParams['figure.dpi'] = 150 #分辨率
plt.figure()
plt.subplot(1, 2, 1)
plt.imshow(mbd)
plt.subplot(1, 2, 2)
plt.imshow(binary_sal)
plt.show()
```

# 图像裁切

```python
from Katna.image import Image as kImage
import cv2
import matplotlib.pyplot as plt

img_path = '../data/image_4_3.jpg'
img = cv2.imread(img_path)

plt.figure()
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
img_module = kImage()
crop_list = img_module.crop_image(file_path=img_path, crop_width=300, crop_height=400, num_of_crops=1)
crop_rect = crop_list[0]

x_pos, y_pos, width, height = crop_rect.x, crop_rect.y, crop_rect.w, crop_rect.h
img = img[y_pos:y_pos + height, x_pos:x_pos + width]

plt.figure()
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.show()
```
